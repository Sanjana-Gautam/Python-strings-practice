```python
# Strings & Examples
# Strings - single double and triple quotes


print("Hello  welcome to the sanna's world !")

print("won't give up")

print(''' "quates" and 'single quates' can be trick ''')

print("\"quates \" and 'single quates ' can be tricky. ")


# formatted string
# old - style formatting (% operator)


name = "sanjana"
age = 20
print("my name is %s and i'm %d " % (name, age))

# # old - style formatting (% operator)+ user input function


name = input("Enter your name ")
age = int(input("Enter your age "))    # %s and %s are placeholder for the sting and int
print("Hii your name is %s and you are %d year old " % (name , age ))


# Formatted String - str.format()


name = "sanna"
age = 20
print("my name is {} and i am {} ".format(name , age))

print("my name is {0} and i am {1}". format(name , age))

print("my name is {name} and i am {age}".format(name = "sanna" , age = 20))


# # Formatted String - str.format() + user input function

name = input("Enter your name")
age = int(input("Enter your age"))
print("my name is {} and i am {} ".format(name , age))

print("my name is {0} and i am {1}". format(name , age))


# Formatted String - F - string

name = "sanjana"
age = 20
print(f"hii my name is {name} and my age is {age} !")

# Formatted String - F - string + user input function

name = input(" Enter your name: ")
age = int(input("Enter your age : "))
print(f"Hii welcome to the sanna's world your name is {name} and your age is {age}")
print(f"Hii {name} welcome to the sanna's world you are just {age} next year you will be {age+1} ")

# Escap Characters

print(" hello \n world ")    # new line
print(" hello \t world ")    # tab
print(" hello \r world ")    # carriage return
print(" hello \b world ")    # backspace
print(" hello \f world ")    #from feed
print(" hello \000 world ")  #Octal values
print(" hello \' world ")    # single quote
print(" hello \" world ")    #double quote
print(" hello \\ world ")    #backslash

# # String operator

a = "hello"
b = "world"

# + concatenation

print(a + b)

# * Repetition

print(a * 2)
print(a * 2 , b * 2)

# [] slice

print(a[0])
print(a[1])
print(a[2])
print(a[3])
print(a[4])

# [ : ] Range Slice

print(a[1:4])

# # membership in  and  not in

if "H" in a :
  print("yess")
else :
  print("noo")

if "h" in a :
  print("Yess")
else :
  print("Noo")

# membership  not in

if "H" not in a :
  print("yess")
else :
  print("noo")


if "h" not in a :
  print("yess")
else :
  print("noo")

# Raw  String

print(r"hello \n world")





























```
